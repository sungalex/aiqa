### KoNLPy 대한민국 법전(kolow) WordCloud

```python
#!pip install konlpy
```

```python
from konlpy.corpus import kolaw
from konlpy.tag import Kkma
```

```python
corpus = kolaw.open(kolaw.fileids()[0]).read()
tagged = Kkma().pos(corpus)
```

```python
#print(corpus)
```

```python
len(tagged)
```

```python
tagged
```

```python
result = [_[0] for _ in tagged if len(_[0]) > 1 and not (_[1].startswith("J") or _[0].startswith("ㄴ"))]    # J* : 조사,  "L"으로 시작하는 종결어미 제외
```

```python
result
```

```python
len(set(result))
```

```python
#!pip install nltk

from nltk import Text
```

```python
textObj = Text(result)
```

```python
textObj.vocab().most_common(20)
```

```python
#!pip install wordcloud
```

```python
from wordcloud import WordCloud
```

```python
termList = textObj.vocab()
```

```python
path = "/Library/Fonts/NanumGothic.ttf"    # for MAC
#path = "C:/Windows/Fonts/malgun.ttf"       # for Windows

wc = WordCloud(font_path=path, max_words=500, background_color="white")
wc.generate_from_frequencies(termList)
wc.to_image()
```

```python
wc_array = wc.to_array()
type(wc_array)
```

### 네이버 뉴스기사 WordCloud

```python
# 네이버 뉴스기사
import os

corpus = ""

for _ in os.listdir("./naver_news"):
    if _.endswith(".txt") and _.startswith("00"):   # 00~05로 시작함(기사 분류 기호)
        with open("./naver_news/"+_, encoding='UTF8') as f:
            corpus += f.read()
```

```python
len(corpus)
```

```python
tagged = Kkma().nouns(corpus)
```

```python
from collections import defaultdict   # Dictionary 자료형
nounList = defaultdict(int)
```

```python
for _ in tagged:
    nounList[_] += 1
```

```python
wc.generate_from_frequencies(nounList)
wc.to_image()
```
